### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

```python
# 1. The Pyplot State-Based Interface:
# This approach relies on the matplotlib.pyplot module, which provides a state-based interface similar to MATLAB.
# You interact with the global state to create and manipulate figures and axes.

# Example:
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [10, 20, 25, 30])
plt.title("State-Based Interface Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()

# Advantages:
# Simple and quick for small scripts and exploratory data analysis.
# Good for one-off plots and basic use cases.
# Disadvantages:
# Less control over multiple figures and axes.
# Harder to manage in complex scripts or large applications.

# 2. The Object-Oriented Interface:
# This approach creates figure and axes objects explicitly and uses their methods to create and customize plots.

# Example:
import matplotlib.pyplot as plt

fig, ax = plt.subplots()  # Create figure and axes
ax.plot([1, 2, 3, 4], [10, 20, 25, 30])  # Plot on the axes
ax.set_title("Object-Oriented Interface Example")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
plt.show()

# Advantages:
# More control over complex layouts, multiple subplots, and figure management.
# Scalable for larger and more complex applications or scripts.
# Encourages better code structure and reusability.
# Disadvantages:
#Slightly more verbose for simple plots.

# Recommended Approach:
# The Object-Oriented Interface is the recommended approach for most use cases, especially when:
# Creating multiple plots or subplots.
# Working on complex visualizations.
# Writing reusable and maintainable code.

# The Pyplot State-Based Interface is better suited for quick, simple plots in interactive environments (e.g., Jupyter notebooks or quick data exploration).
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

# 1. Figure 
The top-level container for all plot elements.
Represents the overall window or page where one or more plots are displayed.
You can think of it as the "canvas" on which everything is drawn.
Created using plt.figure() or implicitly with plt.subplots().

# 2. Axes 
Represents an individual plot or graph within a Figure.
Contains the actual plotting area, including the data, gridlines, labels, and ticks.
A single Figure can have one or multiple Axes (subplots).
Created using fig.add_subplot() or returned by plt.subplots().

# 3. Axis 
Represents the x-axis or y-axis of an Axes object.
Handles the ticks, tick labels, and limits for its respective dimension.
Each Axes object contains two Axis objects (x-axis and y-axis).

# 4. Artist 
The base class for all objects that are drawn on the canvas (everything visible in a plot is an Artist).
Examples include Figure, Axes, Axis, Line2D (for lines), Text, Patch (for shapes), etc.
Artists are responsible for rendering themselves on the canvas.

# Hierarchical Relationship 
Figure contains one or more Axes.
Each Axes contains two Axis objects (x-axis and y-axis).
All visible elements, including the Axes, Axis, lines, labels, etc., are Artists.

### When plotting in Matplotlib, what is the expected input data type?

All of plotting functions expect np.array or np.ma.masked_array as input. Classes that are array-like such as pandas data objects and np.matrix may or may not work as intended. It is best to convert these to np.array objects prior to plotting. 

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
# Object-oriented Approach:

# Define the data
x = np.linspace(-4, 4, 100)  # Generate 100 points between -4 and 4
y = x**2  # Compute y = x^2

# Create the figure and axes
fig, ax = plt.subplots()
ax.plot(x, y, label=r"$y = x^2$")  # Plot the data

# Add title and labels
ax.set_title("Object-Oriented Approach: y = x^2")
ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")

# Display the plot
plt.show()


# Pyplot State-Based Approach:

# Define the data
x = np.linspace(-4, 4, 100)  # Generate 100 points between -4 and 4
y = x**2  # Compute y = x^2

# Create the plot
plt.plot(x, y, label=r"$y = x^2$")

# Add title and labels
plt.title("Pyplot Approach: y = x^2")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

# Display the plot
plt.show()

# Key Points: 
# Title: Both examples include a title. 
# Axis Labels: Both examples label the x-axis and y-axis.
# Range: The np.linspace function ensures the range [4,4] is covered smoothly.
# Customization: Both approaches can include features like legends, line styles, colors, etc.

# Both methods produce identical plots but differ in their structure and coding style.
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

A scatter plot is a type of plot that displays values for typically two variables for a set of data. The values are represented by dots. Matplotlib, a comprehensive library for creating static, animated, and interactive visualizations in Python, provides the scatter() function to create scatter plots. To create a basic scatter plot, you need two arrays of the same length: one for the x-axis values and one for the y-axis values.

A bar plot or bar chart is a graphical representation of data using rectangular bars. The length or height of each bar is proportional to the value it represents. Bar plots can be displayed either horizontally or vertically and are useful for comparing different categories of data. To create a basic bar plot in Matplotlib, you can use the bar() function from the pyplot module.

```python
# Object-Oriented Interface for better control

# Create a figure with 2 subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 5)) 

# plt.subplots(1, 2) creates a single row of two subplots.
# axs[0] and axs[1] are used to control the scatter plot and bar plot, respectively.

# First subplot: Scatter plot: 
axs[0].scatter(random_data_x, random_data_y, alpha=0.5, label='Random Points') # Random points (random_data_x, random_data_y) are plotted using scatter
axs[0].plot(x, y, color='r', label=r"$y = x^2$") # A red curve y=x2 is added to the scatter plot using plot.
axs[0].set_title("Scatter Plot with Curve")
axs[0].set_xlabel("Random X")
axs[0].set_ylabel("Random Y")
axs[0].legend()

# Second subplot: Bar plot
axs[1].bar(names, values, color=['purple', 'green', 'orange', 'yellow', 'lime']) # The fruits' data is used to create a bar plot using bar.
axs[1].set_title("Bar Plot of Fruits") # Each bar represents a fruit and its respective count.
axs[1].set_xlabel("Fruit")
axs[1].set_ylabel("Count")

# Adjust layout and show the figure
fig.tight_layout() # fig.tight_layout() adjusts the spacing to prevent overlap.
plt.show()

# Titles, axis labels, legends, and colors are added for better visualization.
```
